```python
# Pemilihan baris (rows selection) pada Data Frame
n_rows = 10
n_cols = 5
cols= tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)),
                  columns=cols)

df
```
