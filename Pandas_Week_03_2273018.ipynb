```python
# Seleksi kolom dan baris pada Data Frame menggunakan loc
n_rows = 10
n_cols = 5
cols= tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)

df
```

```python
df.loc [[0,3,4], ['B', 'E']]
```

```python
df.loc[df ['B']>10, ['B', 'D', 'E']]
```

```python
# Membentuk kolom bertipe datetime dari beberapa kolom lain pada Pandas Data Frame
data = {'day': [1, 2, 10 ,25, 12],
        'month': [1, 2, 4, 5, 6],
        'year': [2000, 2001, 2018, 2015, 2020]}

df = pd.DataFrame(data)
df
```

```python
df['penaggalan'] = pd.to_datetime(df [['day', 'month', 'year']])
df
```
